```python
# Step 1 - Create four files

# file1=product.txt (open, write, close)
# file2=quantity.txt
# file3=sale_price.txt
# file4=purchase_price.txt

#------------------------------------------#

n=int(input("How many items are there?  ")) # number of data rows to be entered


#------------------------------------------#

#creating the 4 files

product_file = open("product.txt", "w")
quantity_file = open("quantity.txt", "w")
purchaseprice_file = open("purchase_price.txt", "w")
saleprice_file = open("sale_price.txt", "w")


for i in range(n):  #inputting data using for loop

  product=input("What is product number  "   + str(i+1)+ "\n")
  quantity=input("What is quantity of product number  "   + str(i+1) + "\n")
  purprice=input("What is purchase price of product number  "   + str(i+1) + "\n")
  saleprice=input("What is sale price of product number  "   + str(i+1) + "\n")

  product_file.write(product+'\n') # writing input data to files
  quantity_file.write(quantity+'\n')
  purchaseprice_file.write(purprice+'\n')
  saleprice_file.write(saleprice+'\n')

product_file.close(). # closing files
quantity_file.close()
purchaseprice_file.close()
saleprice_file.close()



```

```python
# Revenue = Quantity*Sale_price
# Cost= quantity*purchase_price

revenue=0  #initialising fields for use in calculations
costprice=0
cashback1=3000 # cashback entered here so that its easier to change if needed. no need to change in diferent parts of code if used more than once
cashback2=150
cashback3=0
total_revenue=0
total_cost=0

#replacing $ so that sale and purchase price can be used in calculations

purch_data = open("purchase_price.txt", "r") 
read_purch_data=purch_data.read() #opening file and reading data
read_purch_data=read_purch_data.replace("$"," ") #replacing $ 
purch_data.close() # cloing file
purch_data = open("purchase_price.txt", "w")
purch_data.write(read_purch_data) #opening and re writing without $ sign
purch_data.close()

sale_data = open("sale_price.txt", "r")
read_sale_data=sale_data.read()
read_sale_data=read_sale_data.replace("$"," ")
sale_data.close()
sale_data = open("sale_price.txt", "w")
sale_data.write(read_sale_data)
sale_data.close()

# opening files afer replacing $
prod1 = open("product.txt", "r")
quan1 = open("quantity.txt", "r")
purch1 = open("purchase_price.txt", "r")
sale1 = open("sale_price.txt", "r")



# reading the data in the files with for loop

for quant in quan1:
  saleprice1=sale1.readline()
  purchase1=purch1.readline()
  product1=prod1.readline()

  try: #exception handling to deal with TEN
    
    # calculations for cost price and revenue

    print(product1)
    costprice=int(quant) * int(purchase1)
    print("Costprice is $",costprice)
    revenue=int(quant) * int(saleprice1)
    print("Revenue is $",revenue)
    total_revenue=total_revenue + revenue
    total_cost=total_cost + costprice  


    if revenue > 60000:
     revenue1 = revenue - cashback1
     profit1 = revenue1 - costprice
     print("Thanks!, You have spent more than $60K, you get $3000 Cashback! Total cost for is $", revenue1)
     print("Profit is $", profit1)
     print("-------------------------------------") # to make output easier to read
     

    elif 20000< revenue <= 60000:
      revenue2 = revenue - cashback2
      profit2 = revenue2 - costprice
      print("You spent between $20K and $60K, you get $150 cashback! Total cost is $", revenue2)
      print("Profit is $", profit2)
      print("-------------------------------------")

    
    else: 
      revenue3 = revenue - cashback3
      profit3 = revenue3 - costprice
      print("You spent a total of $",revenue3)
      print("Profit is $", profit3)
      print("-------------------------------------")


  except ValueError:
      print("Error:", quant, "Quantity must be a number")
      print("-------------------------------------")


# final messages re. revenue and profit

print("The total revenue excluding the cashback is $", total_revenue)
overall_profit= total_revenue - total_cost

if overall_profit > 0:  # profit greater than 0
  print("The overall profit is $",overall_profit)

elif overall_profit < 0: # profit less than 0
  print("You did not make any profit")

else: # profit = 0 
  print("you broke even")

print("-------------------------------------")

  




```
